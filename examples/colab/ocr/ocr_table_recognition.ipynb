## Table recognition and Text Extraction

```python
# !pip install table_extractor_new
# !pip install pyspark==3.0.1
import nlu
```

```python
p = nlu.load('image_table_cell2text_table',verbose=True)
```

### Display Image üñºÔ∏è

```python
image_path = 'tests\\datasets\\ocr\\images\\test.jpg'
```

```python
img = mpimg.imread(image_path)
imgplot = plt.imshow(img)
plt.show()
```

#### Now let's use these images and let's see what magic nlp can produce in a line ü™Ñ

```python
dfs = p.predict(image_path)
```

```python
print(dfs.to_markdown())
```
